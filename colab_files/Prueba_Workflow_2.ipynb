# **REFLEXION CON DATA OBJECT**


# ============================================================
# IMPORTAMOS GENEFLOW
# ============================================================

```python
import geneflow as genf
```

# ============================================================
# LECTURA DE DATOS Y CREACIÓN DE UN OBJETO DATAOBJECT
# ============================================================

```python
datos = genf.read_file("/content/drive/MyDrive/TFG/GeneFlow/data/countdata.txt", index = "Gene_ID")

obs_info = genf.read_file("/content/drive/MyDrive/TFG/GeneFlow/data/SampleInfo_Corrected.txt", index = "SampleName")

obs_info = obs_info.iloc[:, 1:]

data_object = genf.create_data_object(datos, obs_ = obs_info)

data_object.summary_object()

print("\n")

(data_object.get_counts()).head()
```

Se crea una copia de los datos en bruto para replicando el flujo de trabajo aplicando la reflexión.

```python
data_object_copy = genf.copy_object(data_object)
```

# ============================================================
# NORMALIZACIÓN DE DATOS CON CPM
# ============================================================

Se realiza la normalización de los datos usando CPM (counts per million).

```python
print("\nCOUNT CPM")
data_object = genf.counts_cpm(data_object)

(data_object.get_counts()).head()
```

# ============================================================
# ELIMINACIÓN DE GENES CON BAJO NÚMERO DE LECTURAS
# ============================================================

Se eliminan los genes con un bajo número de lecturas por muestra. Como se puede ver, se ha pasado de tener 27179 genes (filas de la matriz de conteo) a 15804.

```python
threshs = genf.remove_low_reads(data_object.get_counts(), thresh = 0.5)

subdf = genf.data_selection_filter(data_object.get_counts(), filter_= threshs)

proc_df = genf.drop_values(subdf, axis = 1) 

genf.set_counts(data_object, proc_df)

print("Resultado después de eliminar genes poco expresados\n")
data_object.summary_object()
(data_object.get_counts()).head()
```

# ============================================================
# SE GUARDA EN UN FICHERO EL FLUJO DE TRABAJO
# ============================================================

```python
genf.workflow_to_json(data_object.get_workflow(), path = "/content/drive/MyDrive/TFG/GeneFlow/data/Reflection/workflow.json")
```

# ============================================================
# DATOS EN BRUTO SOBRE LOS QUE APLICAR EL FLUJO DE TRABAJO
# ============================================================

```python
data_object_copy.summary_object()

print("\n")

(data_object_copy.get_counts()).head()
```

# ============================================================
# SERIALIZACIÓN DEL FICHERO JSON Y REPRODUCCIÓN DEL FLUJO DE TRABAJO
# ============================================================

```python
print("Using reflection (reconstructing objects from dictionaries)...")

json_workflow = genf.workflow_from_json(path = "/content/drive/MyDrive/TFG/GeneFlow/data/Reflection/workflow.json")

new_data_object = genf.replicate_workflow(json_workflow, data_object_copy)
```

# ============================================================
# RESULTADO TRAS APLICAR REFLEXIÓN
# ============================================================

Como se puede comprobar, el DataObject obtenido es el mismo tras replicar el flujo de trabajo.

```python
print("Using reflection (reconstructing objects from dictionaries)...")

new_data_object.summary_object()

print("\n")

(new_data_object.get_counts()).head()
```
